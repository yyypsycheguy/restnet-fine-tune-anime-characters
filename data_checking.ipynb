```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os
import torchvision.transforms as transforms
from torchvision.datasets import ImageFolder
```

```python
# load image dataset
# Define how img are transformed
transform = transforms.Compose([
    transforms.Resize((224,224)),
    transforms.ToTensor(),
])

dataset = ImageFolder(root="dataset 2", transform=transform)

# extract a sample image
img_index=1
img, label = dataset[img_index]
img_np = img.numpy()
img_np = img_np.transpose((1,2,0)) # (height, width, channels)

# Extract file names as labels
file_name = os.listdir("dataset 2/abigail_williams_(fate)")

plt.figure(figsize=(20,20))
plt.imshow(img_np)
plt.title(f"index:{label}, label:{file_name}")
plt.show()
```
